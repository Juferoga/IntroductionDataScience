## 2) Limpiando el .csv

Se limpiara el .csv utilizando las funciones y comentando el cambio.
- **c_c19.isnull()**

```python
#print(c_c19.isnull())
c_c19.isnull()
#print(c_c19)
```

El resultado arrojado por la función es **mostrar que campo esta lleno y cual no**, al que esta lleno se le asigna un booleano(*True*) en caso contrario se le asigna un *False* como **indicador** de que se encuentra vacio.

## AXIS

![Axis example](https://railsware.com/blog/wp-content/uploads/2018/11/data-frame-axes.png)

- **c_c19.dropna(axis=0,how='any')**

```python
#print(c_c19.dropna(axis=0,how='any'))
c_c19.dropna(axis=0,how='any')
#print(c_c19)
```

El resultado arrojado por la función es **Elimina todas las filas donde halla un campo que no esta lleno**, el que esta lleno es el unico que se logra ver. Que este dentro del eje "y".
<hr>

- **c_c19.dropna(axis=0,how='all')**

```python
#print(c_c19.dropna(axis=0,how='all'))
c_c19.dropna(axis=0,how='all')
#print(c_c19)
```

El resultado arrojado por la función es **Eliminar el campo que no esta lleno**, el que esta lleno es el unico que se logra ver, en caso contrario no se muestra en la tabla resultante. Dentro del eje "X".
**how=alll** quiere decir todos los registros
<hr>

- **c_c19.dropna(axis=1,how='any')**

```python
#print(c_c19.dropna(axis=1,how='any'))
c_c19.dropna(axis=1,how='any')
#print(c_c19)
```

El resultado arrojado por la función es **Eliminar la columna donde halla un campo que no esta lleno**, el que esta lleno es el unico que se logra ver.

Limpia (axis=1) la columna para cualquier valor que este vacio (how='any')
<hr>

- **c_c19.fillna(value=0)**

```python
#print(c_c19.fillna(value=0))
c_c19.fillna(value=0)
#print(c_c19)
```

El resultado arrojado por la función es el **cambio de los valores NaN por un valor determinado**.
<hr>

- **c_c19.fillna(c_c19.mean())**

```python
#print(c_c19.fillna(c_c19.mean()))
c_c19.fillna(c_c19.mean())
#print(c_c19)
```

El resultado arrojado por la función es el **cambio de los valores NaN por un valor determinado**.
<hr>

- **c_c19.fillna(method='bfill',limit=1)**

```python
#print(c_c19.fillna(method='bfill',limit=1))
c_c19.fillna(method='bfill',limit=1)
#print(c_c19)
```

El resultado arrojado por la función es el **Completa los campos vacios, con el valor por debajo del campo vacío** y solo remplazara el primer campo que encuentre vacio por el método limit=1.

## 2) Filtrado y consulta de datos

Utilizar cada filtro de consulta de datos, y escribir lo que hacen.
- **c_c19['atención'] >= 12**

#### Consultas

Realizar consultas de datos empleando
- **where ...**

```python
Num = pd.read_csv('Numero_de_Habitantes_por_Barrios_Rovira.csv')
print ('\n csv importado x2 👍')
#print(Num)

Num.where(Num > 64)
```

El resultado de la **función es reemplazar con NaN que cumplan con las condiciones dadas**.
<hr>

- **MASK ...**

```python
Num.mask(Num > 64)
```

El resultado de la **función es el contrario de la funcion where**.
<hr>

- **IsIn ...**

```python
Num.isin([2,3,56,98])
```

El resultado de la **función es traer todos los registros y asisgnar verdadero si se cumple que los valores del DF estan en la lista que suministramos**.
<hr>

- **Query ...**

```python
Num.columns = Num.columns.str.replace(" ", "_") # busqueda más simple con _ en vez de " "
query = list(Num)[0] + ">" +list(Num)[1]
print(query)
Num.query('NUMERO_DE_HABITANTES > GENERO_MASCULINO')
```

La función Query trae un **dataFrame con los registros que cumplan la condición del query**.<hr>

- **Eval ...**

```python
print(query)
Num.eval('NUMERO_DE_HABITANTES-GENERO_MASCULINO')
```

La función Eval trae un **dataFrame con los registros elimnados de las columnas que pedimos y muestra su valor en una nueva columna**.<hr>
