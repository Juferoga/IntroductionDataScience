````python
# üìö Librerias necesarias ...

import pandas as pd
import numpy as np
import math

print("Pandas üêº Numpy üá≥ and Math üßÆ imported succesfully!!")
````

````python
# Funciones Utiles

def fctn1(dataframe, col):
    dataframe[col] = np.mean(dataframe[col])
    return dataframe

def fctn2(dataframe, col):
    dataframe[col] = dataframe[col].max()**23
    return dataframe

def fctn3(dataframe, col, num1, num2):
    dataframe[col] = (np.mean(dataframe[col]))*(num1//num2)
    return dataframe

def convert_e(x):
    return x*4.146

def convert_d(x):
    return x*3.730

def convert_l(x):
    return x*4.605
````

````python
# Creando el DataFrame

d = {'w':pd.Series([66,57,75,44,31,67,85,33,42,62,51,47]),
   'h':pd.Series([1.89,1.87,1.67,1.55,1.47,1.72,1.76,1.79,1.44,1.92,1.93,1.69])}
 
df = pd.DataFrame(d)
imc = pd.DataFrame(d)

print (df)
````

````python
print (df.pipe(fctn1,'w'))
````

````python
print (df.pipe(fctn2,'h'))
````

````python
print (df.pipe( fctn3, 'w', 2, 100))
````

````python
#2 

print(imc.apply(lambda x: x['w']/x['h']**2, axis=1))
````

````python
#3

a = np.random.randint(low=0, high=100, size=(5, 3))
df1 = pd.DataFrame(data=a, columns=['A', 'B', 'C'])

## https://blog.adrianistan.eu/estadistica-python-media-mediana-varianza-percentiles-parte-iii ##
print(df1.agg(['median', 'mean', 'sum', 'quantile','skew']))
````

````python
# 4 

# Genere un dataframe donde se almacene el precio (en pesos colombianos) de algunos productos a trav√©s de los 
# meses de un a√±o. A partir del dataframe anterior genere un nuevo dataframe donde los precios de los productos 
# se encuentren en dolares, euros y libras.

price = np.random.randint(low=10000, high= 200000,size=(12,5))
items = np.array(['cadena','mochila','antifaz','tren','guantes'])
dfv = pd.DataFrame(data = price, index = np.arange(1,13), columns = items)
dfv
````

````python
dfv.transform([convert_e,convert_d,convert_l])
````

````python
# 5

solodolares = lambda x: x*3.730
dfv.applymap(solodolares)
````
