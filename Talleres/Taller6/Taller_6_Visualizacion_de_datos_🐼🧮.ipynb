````python
# üìö Librerias necesarias ...

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Pandas üêº Numpy üá≥ and MatplotLib üßÆ imported succesfully!!")
````

````python
# Creando el DataFrame üõ•Ô∏è
 
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Datasets/train.csv')
#print (df)
print()
print("DataFrame imported - Titanic csv üõ•Ô∏è")
print()
df
````

````python
# plot

df['Survived'].cumsum().plot()
````

````python
# bar

df.Embarked.value_counts().plot(kind = 'bar', color='y')
````

````python
# area
df['Embarked'].value_counts().plot.area(stacked=False,colormap='hot')
df.plot.area(stacked=False,colormap='plasma')
````

````python
# hist
#df.hist()#General
df['Age'].hist(bins=10)
````

````python
# box
df['SibSp'].plot.box()
````

````python
# pie
df['Pclass'].value_counts().plot.pie()
````

````python
# table
df['Pclass'].value_counts().plot.pie(table=True,colormap='prism')
#fig, ax = plt.subplots(1,1)
#fd = pd.DataFrame(np.random.randint(1,10),columns=['a','b','c'])
#tb = pd.DataFrame(np.random.rand(18,3),columns=['a','b','c'])
#ax.get_xaxis().set_visible(False)
#fd.plot(table=tb, ax=ax)
#df.plot.table(row='Pclass',columns='Survived')
````
