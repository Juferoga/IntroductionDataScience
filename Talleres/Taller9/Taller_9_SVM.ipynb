## Procedimiento 1 - RBF

````python
x_train, x_test, y_train, y_test = train_test_split(df1.iloc[:,:20], df1.iloc[:,20:21])

y_train
y_test

clf = SVC(gamma=0.001, C=100)

clf.fit(x_train, y_train)

clf.score(x_test, y_test)
print(clf.score(x_test, y_test))

clf.predict(x_test[:-1])
````

## Opinión pros y contras

### Contras

*   Es difícil predecir el modelo final, ya que puede haber muchos cambios minuciosos. Entonces, recalibrar el modelo cada vez no es una solución.
*   Si hay más características que muestras en los datos, el modelo tendrá un bajo rendimiento.
*   SVM toma mucho tiempo para entrenar con grandes conjuntos de datos.
*   El ajuste de los parámetros SVM no es fácil. Su efecto en el modelo es difícil de ver.
*   Requiere mucha memoria.
*   Difícil de interpretar

### Pros

*    alta precisión
*    buenas garantías teóricas con respecto al sobreajuste
*    sin requisito de distribución
*    selección flexible de núcleos para correlación no lineal
*    Incluso si los datos de entrada no son lineales ni separables, los SVM generan resultados de clasificación precisos debido a su solidez.
*    En la función de decisión, utiliza un subconjunto de puntos de entrenamiento llamados vectores de soporte, por lo tanto, es eficiente en memoria.
*    Es útil para resolver cualquier problema complejo con una función de núcleo adecuada.
*    En la práctica, los modelos SVM son generalizados, con menos riesgo de sobreajuste en SVM.
*    Los SVM funcionan muy bien para la clasificación de texto y para encontrar el mejor separador lineal.
*    Las SVM tienen mejores resultados en producción que las ANN.
*    Pueden manejar eficientemente datos de dimensiones superiores e inseparables linealmente. Son bastante eficientes en memoria.
*    Los problemas complejos se pueden resolver utilizando las funciones del núcleo en la SVM. Esto viene bajo el truco del kernel, que es un gran activo para SVM.
*    Podemos agregar diferentes funciones del núcleo juntas para lograr hiperplanos más complejos.
