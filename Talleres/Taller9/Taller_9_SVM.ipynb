## Opini贸n pros y contras

### Contras

*   Es dif铆cil predecir el modelo final, ya que puede haber muchos cambios minuciosos. Entonces, recalibrar el modelo cada vez no es una soluci贸n.
*   Si hay m谩s caracter铆sticas que muestras en los datos, el modelo tendr谩 un bajo rendimiento.
*   SVM toma mucho tiempo para entrenar con grandes conjuntos de datos.
*   El ajuste de los par谩metros SVM no es f谩cil. Su efecto en el modelo es dif铆cil de ver.
*   Requiere mucha memoria.
*   Dif铆cil de interpretar

### Pros

*    alta precisi贸n
*    buenas garant铆as te贸ricas con respecto al sobreajuste
*    sin requisito de distribuci贸n
*    selecci贸n flexible de n煤cleos para correlaci贸n no lineal
*    Incluso si los datos de entrada no son lineales ni separables, los SVM generan resultados de clasificaci贸n precisos debido a su solidez.
*    En la funci贸n de decisi贸n, utiliza un subconjunto de puntos de entrenamiento llamados vectores de soporte, por lo tanto, es eficiente en memoria.
*    Es 煤til para resolver cualquier problema complejo con una funci贸n de n煤cleo adecuada.
*    En la pr谩ctica, los modelos SVM son generalizados, con menos riesgo de sobreajuste en SVM.
*    Los SVM funcionan muy bien para la clasificaci贸n de texto y para encontrar el mejor separador lineal.
*    Las SVM tienen mejores resultados en producci贸n que las ANN.
*    Pueden manejar eficientemente datos de dimensiones superiores e inseparables linealmente. Son bastante eficientes en memoria.
*    Los problemas complejos se pueden resolver utilizando las funciones del n煤cleo en la SVM. Esto viene bajo el truco del kernel, que es un gran activo para SVM.
*    Podemos agregar diferentes funciones del n煤cleo juntas para lograr hiperplanos m谩s complejos.
